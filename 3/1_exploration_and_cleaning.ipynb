```python
import pandas as pd

df = pd.read_csv("healthcare-dataset-stroke-data.csv")

df.head()
```

```python
df.info()
```

```python
df[df["bmi"].isna()]
```

```python
df.loc[df["bmi"].isna(), "bmi"] = df["bmi"].mean()
```

```python
df.iloc[1]["bmi"]
```

```python
df.info()
```

```python
df.head()
```

```python
df["gender"].unique()
```

```python
dummies_gender = pd.get_dummies(df["gender"], drop_first=True, prefix="gender")
dummies_gender
```

```python
df = df.drop("gender", axis=1)
df = df.join(dummies_gender)
df
```

```python
df["ever_married"].unique()
```

```python
df["ever_married"] = (df["ever_married"] == "Yes").astype(int)
```

```python
df["ever_married"].unique()
```

```python
df["work_type"].unique()
```

```python
dummies_work_type = pd.get_dummies(df["work_type"], drop_first=True, prefix="work_type")
df = df.drop("work_type", axis=1)
df = df.join(dummies_work_type)
df
```

```python
df["smoking_status"].unique()
```

```python
dummies_smoking_status = pd.get_dummies(df["smoking_status"], drop_first=True, prefix="smoking_status")
df = df.drop("smoking_status", axis=1)
df = df.join(dummies_smoking_status)
df
```

```python
df.info()
```

```python
dummies_residence_type = pd.get_dummies(df["Residence_type"], drop_first=True, prefix="residence_type")
df = df.drop("Residence_type", axis=1)
df = df.join(dummies_residence_type)
df
```

```python
df = df.drop("id", axis=1)
df.head()
```

```python
df.to_csv("healthcare-dataset-stroke-data-clean.csv", index=False)
```
